```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import random
```

```python
rentals = pd.read_csv("https://raw.githubusercontent.com/Codecademy/datasets/master/streeteasy/streeteasy.csv")
```

```python
rentals.head()
```

```python
rentals.no_fee.replace({1:'yes', 0:'no'}, inplace=True)
rentals.has_roofdeck.replace({1:'yes', 0:'no'}, inplace=True)
rentals.has_washer_dryer.replace({1:'yes', 0:'no'}, inplace=True)
rentals.has_doorman.replace({1:'yes', 0:'no'}, inplace=True)
rentals.has_elevator.replace({1:'yes', 0:'no'}, inplace=True)
rentals.has_dishwasher.replace({1:'yes', 0:'no'}, inplace=True)
rentals.has_patio.replace({1:'yes', 0:'no'}, inplace=True)
rentals.has_gym.replace({1:'yes', 0:'no'}, inplace=True)
rentals.drop(columns=['rental_id', 'building_id'], inplace=True)
```

## Analyse the association between one continuous quantitative variable and one categorical variable

```python
# break down thinking about associations into the types of variables
# /sets based on the types of variables that you're trying to find an
# association between

# in the apartment dataset, we have quantitative variables (rent, bedrooms, floor)
# we also have categorical variables (borough, has_gym, has_patio)

# start with rent vs borough
# calculate summary statistic of rent in each borough

# df.groupby() function is basically you saying: I want to calculate 
# some statistic for particular groups within my data
rentals.groupby(by='borough').mean()
# nice way to quickly pull out summary statistics by borough

# get just the mean rent for each borough
rentals.groupby(by='borough').mean()['rent']
# looks like Queens might be cheapest to live in
```

```python
# calculate the median rent for each borough
rentals.groupby(by='borough').median()['rent']

# notice that the mean is bigger than the median for each borough
# right skewed
```

```python
# let's create a boxplot - simplest of all
plt.figure(figsize=(10,6))
sns.boxplot(x='borough', y='rent', data=rentals)

# notice: the outliers are squishing the boxes
```

```python
# modify the data a bit and create a boxplot again
plt.figure(figsize=(10,6))
sns.boxplot(x='borough', y='rent', data=rentals[rentals['rent'] < 12000])

# the instructor checked whether the boxes overlap
# the fact that the boxes don't overlap much at all makes it kinda clear
# that what we're seeing are pretty highly associated variables
# i.e. if we know what borough the apartment is in, we can determine
# what the rent is more or less likely to be
```

```python
# checking association between rent and washer_dryer

plt.figure(figsize=(10,6))
sns.boxplot(x='has_washer_dryer', y='rent', data=rentals)

# so the boxes are closer than I imagined
# doesn't look like the median is that different
```

```python
# quickly checking the difference in mean between the two answers above
rentals.groupby(by='has_washer_dryer').median()['rent']
```

```python
# let's ask a question
# is the price lower for a high floor with no elevator

plt.figure(figsize=(10,6))
sns.boxplot(x='has_elevator', y='rent', data=rentals[rentals['floor'] < 30])
plt.title('Is there a rent difference between places with less than 30 floors that do or do not have elevators?')
plt.show()
# rentals[rentals['floor'] < 30]
# rentals['floor'].unique() # checking the range of floors

# Analysis: If the building has an elevator, the median price of the room
# is slightly higher than one in a building with no elevator
# there is more rent variation among rooms in buildings with elevators
```

```python
# another great visualisation for seeing an association between 
# a quantitative and a categorical variable is an overlaid histogram

# plt.hist(rentals[rentals['borough'] == 'Manhattan'], label='Manhattan', bins=30)



plt.hist(rentals.loc[rentals['borough']=='Queens', 'rent'], label='Queens', density=True, bins=30, alpha=.5)
plt.hist(rentals.loc[rentals['borough']=='Manhattan', 'rent'], label='Manhattan', density=True, bins=30, alpha=.5)
plt.hist(rentals.loc[rentals['borough']=='Brooklyn', 'rent'], label='Brooklyn', density=True, bins=30, alpha=.5)
plt.legend()
plt.show()

# Notice: without the density parameter, the histograms are not of equal size
# that's because there are many times more apartments in Manhattan than Queens

# DENSITY PARAMETER - returns a probability density
# basically, what the density parameter does is force all of the histograms to
# be equal to an area of 1. Previously, we were seeing frequencies. But now
# we are seeing the area; all of the histograms are on the same scale
```

## Analysing two quantitative variables

Enter the scatterplot!

```python
# create a scatterplot to analyse size_sqft and rent

plt.scatter(x=rentals['size_sqft'], y=rentals['rent'])
plt.xlabel('sqft')
plt.ylabel('rent')
plt.show()

# y is the dependent variable

# NOTE: The instructor said she almost always plots the data/visualises it
# to see/tell what's going on before calculating the summary statistic(s)
```

```python
# a summary statistic that can help us understand the scatterplot above
# is the Pearson Correlation, which essentially gives you a number
# between -1 and 1 that tells you how strong of a linear 
# (whether you could draw a line between those two variables or not)
# relationship between two variables

# where larger values of one variable are associated with larger/smaller
# values of another variable

from scipy.stats import pearsonr
pearsonr(x=rentals['size_sqft'], y=rentals['rent'])

# you can see that the correlation is pretty high
# 0.0 is a p-value
```

```python
# another way to explore a correlation is to do a .corr() function
# on the entire dataframe - WOAH

rentals.corr()
# it shows only numerical variables
```

```python
# heatmap is a way to visualise the correlation matrix

# first, set a colour gradient
colours = sns.diverging_palette(h_neg=150, h_pos=275, as_cmap=True)

# then create the heatmap
sns.heatmap(rentals.corr(), center=0, cmap=colours)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
```

```python
# let's explore the 1 negative correlation in the matrix

plt.scatter(x=rentals['building_age_yrs'], y=rentals['floor'])
plt.xlabel('building_age_yrs')
plt.ylabel('floor')

# so it looks like in the 50s and 60s, developers built max
# 20 storey buildings in Manhattan, Brooklyn, and Queens;
# most buildings built in that era seem to be about 10 storeys high

# if the building is newer, there's a much wider range of floors

# so the scatterplot is a mess of values
```

```python
# let's explore 0 correlation

plt.scatter(x=rentals['size_sqft'], y=rentals['min_to_subway'])
plt.xlabel('size_sqft')
plt.ylabel('min_to_subway')

# almost a slight negative correlation but not really
```

```python
# quickly visualise all of the correlations/associations
# using seaborn's pairplot function

sns.pairplot(rentals)
plt.show()
```

## Analysing two categorical variables

```python
# has_elevator vs has_gym

# use a two-way table or a cross-tabulation 
# this is basically a grid showing all possible combinations
# of categories for a categorical variable and in the grid
# it will show frequencies or proportions

# used Pandas crosstab function

pd.crosstab(index=rentals['has_elevator'], columns=rentals['has_gym'])

# looking at the table below, we can see that if an apartment has no elevator,
# it's highly unlikely for it to have a gym as well

# but for apartments that do have an elevator, the gym-to-no-gym ratio is
# more evenly split
```

```python
# let's visualise the table above
# use seaborn's countplot

sns.countplot(x='has_elevator', hue='has_gym', data=rentals)
# countplot is really a barplot

# analysis: if it doesn't have an elevator, it almost definitely doesn't
# have a gym
```

```python
# testing a theory about whether or not two variables are related at
# a population level

# we can use the chi2_contingency function from the scipy.stats module
# Chi-square test of independence of variables in a contingency table. 
# This function computes the chi-square statistic and p-value for the 
# hypothesis test of independence of the observed frequencies in the 
# contingency table [1] observed

# import the function
from scipy.stats import chi2_contingency
xtab_elevator_gym = pd.crosstab(index=rentals['has_elevator'], columns=rentals['has_gym'])

# run the function
chi2_contingency(observed=xtab_elevator_gym)

# we get an output that is a matrix
# what we get is a matrix of expected values that would exist here
# if there was no relationship between these two variables

# the table below is what we'd see if the null hypothesis is true
```
